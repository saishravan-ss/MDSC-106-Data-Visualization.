```python
def HistPlot(df):
    df=merged_df
    fig,axes = plt.subplots(3,2,figsize=(10,7))
    a,b = 0,0
    numerical_features = ['Age', 'CustomerSince', 'HighestSpend', 'MonthlyAverageSpend', 'Mortgage']

    for cols in merged_df.columns:
        if cols in numerical_features:
            sns.histplot(merged_df[cols],kde=True,ax=axes[a,b])
            axes[a,b].set_title(f'Plot for {cols}')
            b+=1
            if b>=2:
                b=0
                a+=1
                if a >= 3:
                    break 
HistPlot(merged_df)
plt.show()
```

'Age', 'CustomerSince', 'HighestSpend', 'MonthlyAverageSpend', 'Mortgage' -- are all continuous numerical features where we can
plot the histograms for these features and can determine their central tendency and spread of these variables.

We can see that plots for HighestSpend and MonthlyAverageSpend are both right-skewed, and the both are getting gradually decreased.And we can see that there are no mortgages done and very few mortgages can be seen.

Seems like the distribution is normal for the both the features Age and CustomerSince.

```python
def BarPlot(df):
    df = merged_df
    figure,axes = plt.subplots(1,3,figsize=(10,5))
    c,d = 0,0
    colu = ['Level','ZipCode']
    for col in merged_df.columns:
        if col in colu:
            sns.barplot(merged_df[col],ax=axes[d])
            axes[c].set_title(f'Plot for {cols}')

BarPlot(merged_df)
plt.show()
```

```python
def Bar(df):
    df = merged_df
    figure, axes = plt.subplots(2,3,figsize=(15,5))
    a,b = 0,0
    features = ['FixedDepositAccount','InternetBanking','CreditCard','LoanOnCard']
    for cols in merged_df.columns:
        if cols in features:
            sns.barplot(merged_df[cols],ax=axes[a,b])
            axes[a,b].set_title(f'Bar Plot for {cols}')
            b+=1
            if b>=3:
                b=0
                a+=1
                if a >= 2:
                    break 

Bar(merged_df)                    
plt.show()

```

```python
def Pie(df):
    df = merged_df
    categorical_columns = ['Level', 'FixedDepositAccount', 'InternetBanking', 'CreditCard', 'LoanOnCard']
    fig, axes = plt.subplots(1, len(categorical_columns), figsize=(15, 5))
    for i, column in enumerate(categorical_columns):
        category_counts = merged_df[column].value_counts()
        labels = category_counts.index
        sizes = category_counts.values
        axes[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
        axes[i].set_title(f'Distribution of {column}')
        axes[i].axis('equal')  # Ensure the pie chart is drawn as a circle

Pie(merged_df)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def pie_chart_matrix(dataframe, figsize=(12, 12)):
    features = merged_df.select_dtypes(include=['object', 'category', 'bool'])
    num_features = len(features.columns)
    num_cols = int(num_features**0.5)
    num_rows = (num_features + num_cols - 1) // num_cols
    fig, axes = plt.subplots(num_rows, num_cols, figsize=figsize)
    axes = axes.ravel()
    
    for i, feature in enumerate(features.columns):
        counts = merged_df[feature].value_counts()
        labels = counts.index
        sizes = counts.values
        axes[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
        axes[i].set_title(f'Distribution of {feature}')
        axes[i].axis('equal')
    for i in range(num_features, num_cols * num_rows):
        fig.delaxes(axes[i])
    
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

def hist_plot(dataframe, figsize=(15,15)):
    features = merged_df.select_dtypes(include=['number'])
    num_features = len(features.columns)
    num_cols = int(math.sqrt(num_features))
    num_rows = math.ceil(num_features / num_cols)
    fig, axes = plt.subplots(num_rows, num_cols, figsize=figsize)
    axes = axes.ravel()
    for i, feature in enumerate(features.columns):
        axes[i].hist(merged_df[feature], bins=20, alpha=0.7, color='pink', edgecolor='black')
        axes[i].set_title(f'Histogram of {feature}')
    for i in range(num_features, num_cols * num_rows):
        fig.delaxes(axes[i])
    
    plt.tight_layout()
    plt.show()
```
