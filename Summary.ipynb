1)  Introduction to Pandas :

Pandas is an open-source data manipulation and analysis library in python.
Two key components : Series and DataFrames.

Dataframes - given the data file it arranges the data in rows and columns 
Series - given the data file it arranges the data in only columns.

DataFrames can be easily manipulated, cleaned, and analyzed.

Import pandas as pd – we import pandas in this way when we use them

Reading csv: df = pd.read_csv(‘data.csv’) – read the csv file

Functions:
→ df.head(n) : gives us the first n rows of the dataframe.
→ df.tail(n) : gives us the last n rows of the dataframe
→ df.shape : gives us the no.of rows and columns.
→ df.info() : gives us the summary of the dataframe it has.
→ df.isnull() : gives us the null values.
→ del df[‘column_name’] : deletes the particular column in the dataframe
→ df.fillna(value) - fills the missing value 
→ df.drop_duplicates(inplace=True) - removes the duplicate values in the dataframe

Accessing : 
df[‘column_name’] - gives the entire column
df.loc[row_index] - gives the particular row when mentioned index.
Df.at[row_index,’column_name’] = give the particular element in that row and at that column.

df[‘new_column’] : adds the new column to the dataframe
df.drop[‘column_name’,axis=1,inplace = True] : drops the column given
Filtered_df = df[df[‘’Age’]>25] : filters the data in the age column which is above 25

```python
import pandas as pd
import seaborn as sns
from sklearn import datasets as ds
import matplotlib.pyplot as plt

data_dict = ds.load_diabetes(as_frame=True)
df = data_dict['data']
data_dict['target']
df['target'] = data_dict['target']
df
```

```python
plt.figure(figsize=(8,6))
plt.subplot(1,4,1)
sns.histplot(df['age'],kde=True,color='pink')
plt.title('Histogram')
plt.xlabel('Age')

plt.subplot(1,4,2)
sns.scatterplot(x='age',y='bp',data=df,hue='sex',palette='deep')
# plt.scatter(df['age'],df['bmi'],data=df,color='r')
plt.title('Scatterplot')
plt.xlabel('Age')
plt.ylabel('bp')

plt.subplot(1,4,3)
sns.boxplot(y=df['target'],color='r')
plt.title('Boxplot')
plt.ylabel('Target')

plt.subplot(1,4,4)
sns.barplot(x='target',y='bmi',data = df)
plt.title('Barplot')
plt.xlabel('Target')
plt.ylabel('BMI')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
data1 = np.random.normal(100,10,200)
data2 = np.random.normal(90,20,200)
data3 = np.random.normal(80,30,200)
data4 = np.random.normal(70,40,200)
data = [data1,data2,data3,data4]

fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
bp = ax.boxplot(data)
plt.title('Box Plot')
plt.show()
```

Box Plotssummarizes data values.
In box plot a box is created from the first quartile to the third quartile a vertical line is also there which goes through the box at the median.

X-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

Stacked bar -- Plots the bars inthe stack manner.

```python
x = ['A','B','C','D']
y1 = [10,20,10,40]
y2 = [20,25,15,25]

plt.bar(x,y1,color='pink')
plt.bar(x,y2,bottom = y1,color='green')
plt.show()
```

```python
import pandas as pd
data = {
    "Name":['Sai','Shravan','Gade'],
    "Age" : [21,21,21],         # when we give less values than the required amount it gives the error
    "Place" : ['Nizamabad','Hyderabad','Telangana'],
}

df = pd.DataFrame(data)
df
```

```python
print(df.head(1)) # gives the first value in the DataFrame
print()
print(df.tail(1))   # gives the last value in the DataFrame
```

```python
print(df['Name']) # gives full coloumn
print()
print(df.loc[1]) # gives that row information
print()
print(df.at[1,'Name']) # gives the elemnet at the index and coloumn mentioned
print()
```

```python
df.isnull()  # used for checking whether null values
```
